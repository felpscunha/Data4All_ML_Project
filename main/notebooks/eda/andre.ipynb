This data set presents an employee survey from IBM, indicating if there is employee attrition or not. The data set contains approximately ~1200 entries. 

Data Description

Id - an anonymous id given to an Employee ( Identificador do funcionário)

Age - Age of an Employee (Idade do funcionário)

Attrition - Did the Employee leave the company, 0-No, 1-Yes (O funcionário saiu da empresa? 0 - Não, 1 - Sim)

BusinessTravel - Travlling frequency of an Employee ( Frequência de viagens de negócio)

Department - Work Department (Departamento)

DistanceFromHome - Distance of office from home (Distância para a residência)

EducationField - Field of Education (Área de formação)

EmployeeNumber - Number of Employees in the division of a given Employee (Número de funcionários em uma divisão de dado funcionário (????)

EnvironmentSatisfaction - Work Environment Satisfaction (Satisfação com o ambiente de trabalho)

Gender - Gender of Employee (Gênero)

MartialStatus - Martial Status of an employee (Estado civil) 

MonthlyIncome - Monthly Income of Employee in USD (Salário em dólares americanos)

NumCompaniesWorked - Number of Companies in which Employee has worked before joining this Company (Número de empresas que o funcionário trabalhou antes de se juntar a empresa atual)

OverTime - Does The person work overtime ( Se o funcionário faz hora-extra)

PercentSalaryHike - Average annual salary hike in percentages (Porcentagem anual média de aumento salarial)

StockOptionLevel - Company stocks given to an Employee (Ações dadas aos funcionários)

TotalWorkingYears - Total working experience of an employee ( Tempo total de experiência do funcionário)

TrainingTimesLastYear - No. of trainings an employee went through last year (Número de treinamentos realizados no último ano)

YearsAtCompany - Number of years worked at this company (Quantidade de anos trabalhados na empresa)

YearsInCurrentRole - Number of years in current role (Quantidade de anos na função atual)

YearsSinceLastPromotion - Number of years since last promotion (Quantidade de anos desde a última promoção)

YearsWithCurrManager - Number of years with the current manager (Quantidades de ano com o gestor atual)

```python
# Mantendo uma cópia para a análise: 

df['BusinessTravelClass'] = df['BusinessTravel']

#Label enconding da variável 'BusinessTravelClass'.

label_business_travel = {'Non-Travel' : 0,
                         'Travel_Rarely' : 1,
                         'Travel_Frequently' : 2}
df['BusinessTravel'] = df['BusinessTravel'].map(label_business_travel)
df.head()
```

```python
df['Department'].unique()
```

```python
# Mantendo uma cópia para a análise: 

df['DepartmentName'] = df['Department']

# One hot encoding para 'Departaments' 

df = pd.get_dummies(df, columns = ['Department'],  dtype = 'int')

df.head()
```

```python
df['EducationField'].unique()
```

```python
# Mantendo uma cópia para a análise: 

df['EducationFieldName'] = df['EducationField']

# One hot encoding para 'Departaments' 

df = pd.get_dummies(df, columns = ['EducationField'],  dtype = 'int')

df.head()
```

```python
df['Gender'].unique()
```

```python
# Mantendo uma cópia para a análise: 

df['GenderName'] = df['Gender']

#Label enconding da variável 'BusinessTravelClass'.
label_binary_gender = {'Female' : 0,
                 'Male' : 1}
df['Gender'] = df['Gender'].map(label_binary_gender)
df.head()
```

```python
df['JobRole'].value_counts()
```

```python
# Mantendo uma cópia para a análise: 

df['JobRoleName'] = df['JobRole']

#One hot enconding da variável 'JobRole'.

df = pd.get_dummies(df, columns = ['JobRoleName'],  dtype = 'int')

df.head()
```

```python
df['MaritalStatus'].unique()
```

```python
# Mantendo uma cópia para a análise: 

df['MaritalStatusName'] = df['MaritalStatus']

#One hot enconding da variável 'MaritalStatus'.

df = pd.get_dummies(df, columns = ['MaritalStatus'],  dtype = 'int')

df.head()
```
